```python
import numpy as np
import pandas as pd  
import urllib.request
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
res = urllib.request.urlopen(url)

html = res.read()

## parse with BeautifulSoup
bs = BeautifulSoup(html, "html.parser")

tables = bs.find_all("table")

df = (pd.read_html(str(tables)))[0]

columns=["Postcode", "Borough", "Neighbourhood"]
df.columns=columns
print(df)
```

```python
df.drop(df[df['Borough']=="Not assigned"].index,axis=0, inplace=True)
df.head()
```

```python
df.info
```

```python
df_grp= df.groupby('Postcode').agg(lambda x: ','.join(x))
df_grp.loc[df_grp['Neighbourhood']=="Not assigned",'Neighbourhood']=df_grp.loc[df_grp['Neighbourhood']=="Not assigned",'Borough']
df_grp['Borough']= df_grp['Borough'].str.replace('nan|[{}\s]','').str.split(',').apply(set).str.join(',').str.strip(',').str.replace(",{2,}",",")
df_grp.head()
```
